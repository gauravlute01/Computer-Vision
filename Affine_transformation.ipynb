Affine Transformation

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# import image 
img = cv2.imread('lunar_input.jpg')
rows, cols = img.shape[:2]
```

```python
plt.imshow(img)
```

```python
tx, ty = 50, 30
M_translation = np.float32([[1,0, tx], [0,1, ty]])
```

```python
translated = cv2.warpAffine(img, M_translation, (cols, rows))
```

```python
plt.imshow(translated)
```

Rotation

```python
angle = 45
center = (cols//2, rows//2)
scale = 1.0
```

```python
M_rotation = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(img, M_rotation, (cols, rows))
```

```python
plt.imshow(rotated)
```

Scaling

```python
sx, sy = 0.5, 0.5
M_scaling = np.float32([[sx,0, 0], [0, sy, 0]])
scaled = cv2.warpAffine(img, M_scaling, (cols, rows))
```

```python
plt.imshow(scaled)
```

Shearing 

```python
shear_factor = 0.5
M_shear = np.float32([[1,shear_factor, 0], [0,1,0]])
sheared = cv2.warpAffine(img, M_shear, (cols + 100, rows))
```

```python
plt.imshow(sheared)
```

combined Transformation

```python
M_combined = M_translation * M_rotation
transformed = cv2.warpAffine(img, M_combined, (cols, rows))
```

```python
plt.imshow(transformed)
```

Handling Edge Cases

```python
rotated1 = cv2.warpAffine(img, M_rotation, (cols, rows), 
                         borderMode=cv2.BORDER_CONSTANT, 
                         borderValue=(255, 255, 255))
```

```python
plt.imshow(rotated1)
```

```python
# All visualization 
plt.figure(figsize=(12,6))
plt.subplot(2,3,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(2,3,2)
plt.title("Translated image")
plt.imshow(translated)

plt.subplot(2,3,3)
plt.title("Rotation")
plt.imshow(rotated)

plt.subplot(2,3,4)
plt.title("Scaling image")
plt.imshow(scaled)

plt.subplot(2,3,5)
plt.title("Shearing image")
plt.imshow(sheared)

plt.subplot(2,3,6)
plt.title('Handling edge cases')
plt.imshow(rotated1)

plt.tight_layout()
plt.savefig('Affine_transformation_output1.png')
plt.show()
```
