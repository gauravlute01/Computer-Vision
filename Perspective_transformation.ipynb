Perspective Transfprmation 

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

Image 1

```python
# Load image 
image = cv2.imread('lunar_input.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
# Define source points 
src_points = np.float32([[320, 300], [740, 300], [200, 650], [880, 650]])
```

```python
# Define destination point 
dst_points = np.float32([[0,0],[600,0],[0,800],[600, 800]])
```

```python
matrix = cv2.getPerspectiveTransform(src_points, dst_points)
```

```python
warped = cv2.warpPerspective(image, matrix, (600, 800))
```

```python
# Display results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image)
plt.scatter(src_points[:, 0], src_points[:, 1], c='red', s=50)  # Mark source points
plt.legend()

plt.subplot(1, 2, 2)
plt.title("Perspective-Transformed Image")
plt.imshow(warped)

plt.tight_layout()
plt.savefig('Perspective_transformation_output1.png')
plt.show()
```
