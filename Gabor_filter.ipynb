Gabor filter 
- Purpose : edeg detection, texture analysis and feature extraction.

```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
img = cv2.imread('gabor_input1.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
```

```python
# create gabor filter 
def create_gabor_filter(ksize, sigma, theta, lambd, gamma, psi):
    
    gabor_kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lambd, gamma, psi, ktype=cv2.CV_32F)

    return gabor_kernel
```

```python
#  Define the parameters 
ksize = 31
sigma = 3.0
theta = np.pi / 4
lambd = 10
gamma = 2
psi = 0
```

```python
# create gabor kernel 
gabor_kernel = create_gabor_filter(ksize, sigma, theta, lambd, gamma, psi)
```

```python
filtered_image = cv2.filter2D(img, cv2.CV_8UC3, gabor_kernel)
```

```python
plt.imshow(filtered_image, cmap='gray')
```

```python
# visualization 
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('Filtered Image(Gabor)')
plt.imshow(filtered_image, cmap='gray')

plt.savefig('gabor_output1.png')
plt.show()
```
