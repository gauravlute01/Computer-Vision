```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
# Open and display the images 
img = cv2.imread('Road_Original_image.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
# Aplly binary threshold to the image 
img = cv2.imread('Road_Original_image.png', 0)
ret , thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
display_img(thresh1, cmap='gray')
```

```python
# convert HSV colorspace 
img = cv2.imread('Road_Original_image.png')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
display_img(img)
```

```python
# Create a low pass filter with a 4 by 4 kernel filled with values of 1/10 (0.01) and then 2-D convolution to blur.
```

```python
kernel = np.ones(shape=(4,4), dtype=np.float32)
```

```python
kernel
```

```python
img = cv2.imread('Road_Original_image.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
dst = cv2.filter2D(img, -1, kernel)
display_img(dst)
```

```python
# Create a horizontal Sobel filter 
img = cv2.imread('Road_Original_image.png')
```

```python
sobelx = cv2.Sobel(img, cv2.CV_64F,1,0, ksize=5)
```

```python
display_img(sobelx, cmap='gray')
```

```python
# Plot histogram 
img = cv2.imread('Road_Original_image.png')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Road image Histograms')
plt.show()

```
