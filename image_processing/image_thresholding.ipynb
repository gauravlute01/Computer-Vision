Image Thresholding 

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('rainbow.jpeg')
img = cv2.resize(img, (450, 600))
```

```python
plt.imshow(img)
```

```python
# Adding the 0 flag to read it in black and white 
img = cv2.imread('rainbow.jpeg', 0)
```

```python
plt.imshow(img)
```

Different Threshold Types 

Binary 

```python
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
ret 
```

```python
plt.imshow(thresh1, cmap='gray')
```

Binary Inverse

```python
ret, thresh2 = cv2.threshold(img, 127, 155, cv2.THRESH_BINARY_INV)
plt.imshow(thresh2, cmap='gray')
```

Threshold Truncation 

```python
ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
```

```python
plt.imshow(thresh3, cmap='gray')
```

Threshold to Zero 

```python
ret , thresh4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
```

```python
plt.imshow(thresh4, cmap='gray')
```

Threshold to zero (Inverse)

```python
ret, thresh5 =  cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thresh5, cmap='gray')
```

Real world Application

Adaptive Thresholding 

Sudoku Image 

```python
img = cv2.imread('paper.jpg', 0)
plt.imshow(img)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

Simple Binary 

```python
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
show_pic(th1)
```

Adaptive Threshold

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
```

```python
show_pic(th2)
```

```python
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 8)
```

```python
show_pic(th3)
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.7, src2=th2, beta=0.3, gamma=0)
show_pic(blended)
```
