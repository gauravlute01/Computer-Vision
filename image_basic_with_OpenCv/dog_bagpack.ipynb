```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('dog_bagpack.jpeg')
```

```python
# Wrong RGB order 
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Flip image 
new_img = img_rgb
new_img = cv2.flip(new_img, 0)
plt.imshow(new_img)
```

```python
plt.imshow(img_rgb)
```

```python
cv2.rectangle(img_rgb,pt1=(70,50),pt2=(120, 100),color=(0,0,150),thickness=5)
plt.imshow(img_rgb)
```

```python
# Draw blue triangle 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
vertices = np.array([[80,85],[100,40],[120,85]],np.int32)
pts = vertices.reshape((-1,1,2))

cv2.polylines(img_rgb,[pts],isClosed= True,color=(0,0,255),thickness=3)
plt.imshow(img_rgb)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
vertices = np.array([[80,85],[100,40],[120,85]],np.int32)
pts = vertices.reshape((-1,1,2))

cv2.fillPoly(img_rgb, [pts], color=(0,0,255))
plt.imshow(img_rgb)
```
