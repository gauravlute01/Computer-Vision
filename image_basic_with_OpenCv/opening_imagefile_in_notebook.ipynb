```python
# Opening image file in a notebook 
import numpy as np 
import cv2
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('dogs.jpeg')
print(img)
```

```python
img_bgr = cv2.imread('dogs.jpeg')
plt.imshow(img_bgr)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('dogs.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
# Resize image 
img_rgb.shape
```

```python
img = cv2.resize(img_rgb, (1300, 275))
```

```python
plt.imshow(img)
```

```python
# By ratio
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
#%matplotlib qt
plt.imshow(new_img)
```

```python
# Flipping images 
```

```python
# Along central x axis
new_img = cv2.flip(new_img, -1)
plt.imshow(new_img)
```

```python
# Along central y axis 
new_img = cv2.flip(new_img, 1)
plt.imshow(new_img)
```

```python
# saving images files 
type(new_img)
```

```python
cv2.imwrite('my_new_pictures.jpg', new_img)
```

```python
# Larger display in the notebook 
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(new_img)
```
