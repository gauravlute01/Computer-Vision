Color Spaces RGB, HSV

```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
```

Image 1

```python
img = cv2.imread('room_input1.png')
```

```python
# convert color spaces 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,3,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(2,3,2)
plt.title('RGB Image')
plt.imshow(img_rgb)

plt.subplot(2,3,3)
plt.title('HSV Image')
plt.imshow(img_hsv)

plt.subplot(2,3,4)
plt.title('HLS Image')
plt.imshow(img_hls)

plt.subplot(2,3,5)
plt.title('YcrCB Image')
plt.imshow(img_ycrcb)

plt.subplot(2,3,6)
plt.title('Lab Image')
plt.imshow(img_lab)

plt.tight_layout()
plt.savefig('Color_spaces_output1.png')
plt.show()
```
