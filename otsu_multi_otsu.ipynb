Otsu Segmentation

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

Image 1

```python
image = cv2.imread('saturn.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image,cmap='gray')
```

```python
# # Apply otsu Thresholding 
# image, otsu_threshold = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
_, otsu_threshold = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.imshow(otsu_threshold, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image Saturn")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Otsu Segmentation")
plt.imshow(otsu_threshold, cmap='gray')

plt.savefig('otsu_output1.png')
plt.show()
```

Multi Otsu thresholding

Image2

```python
from skimage import io
from skimage.filters import threshold_multiotsu
```

```python
img = io.imread('multi_otsu_input.png', as_gray=True)
```

```python
plt.imshow(img, cmap='gray')
```

```python
# Apply multi-otsu thresholding
thresholds = threshold_multiotsu(img, classes=2)
```

```python
segmented = np.digitize(img, bins=thresholds)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('Multi-otsu segmentation')
plt.imshow(segmented, cmap='gray')

plt.savefig('otsu_Output2.png')
plt.show()
```
