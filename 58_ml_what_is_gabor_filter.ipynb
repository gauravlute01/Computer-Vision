```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
ksize = 5
sigma = 3
theta = 1*np.pi/4
lamda = 1*np.pi/4
gamma = 0.4
phi = 0
```

```python
kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)
```

```python
plt.imshow(kernel)
```

```python
ksize = 50
sigma = 3
theta = 1*np.pi/4
lamda = 1*np.pi/4
gamma = 0.4
phi = 0
kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)
plt.imshow(kernel)
```

```python
ksize = 50
sigma = 30
theta = 1*np.pi/4
lamda = 1*np.pi/4
gamma = 0.4
phi = 0
kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)
plt.imshow(kernel)
```

```python
img = cv2.imread("solar_panle_2.jpg")
#img = cv2.cvtColor(img, cv2.COLOR_BAYER_BG2GRAY)
fimg = cv2.filter2D(img, cv2.CV_8UC3, kernel)
```
