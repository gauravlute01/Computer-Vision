Canny Edge Detection

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('solar_panel_1.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=70, threshold2=70)
```

```python
plt.imshow(edges)
```

```python
edges = -cv2.Canny(image=img, threshold1=0, threshold2=100)
```

```python
plt.imshow(edges)
```

Choosing Threshold

```python
# calculate the median 
med_val = np.median(img)

med_val
```

```python
# Lower bound is 0 or 70% of the median  value, whichever is higher
lower = int(max(0, 0.7*med_val))
lower
```

```python
# Upper bound is either 255 or 30% above the median value, whichever is lower
upper = int(min(255, 1.3*med_val))
upper
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
```

```python
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img, ksize=(5,5))
```

```python
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper)
```

```python
plt.imshow(edges)
```
