Canny Filter

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('Work_canny_input1.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray')
```

```python
# Apply Guassian blur 
blurred = cv2.GaussianBlur(image, (5,5), sigmaX=1.4)
```

```python
# Apply canny edge detection with threshold 
low_threshold = 100
high_threshold = 200
edges = cv2.Canny(blurred, threshold1=low_threshold, threshold2=high_threshold, L2gradient=True)
```

```python
# visualization
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1,2,2)
plt.title('Canny Edge')
plt.imshow(edges, cmap='gray')

plt.savefig('canny_output1.png')
plt.show()
```
