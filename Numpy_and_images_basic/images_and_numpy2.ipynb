```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt 
```

```python
from PIL  import Image 
```

```python
pic = Image.open('dogs.jpeg')
pic
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
pic_arr.shape
```

```python
plt.imshow(pic)
```

```python
pic_red = pic_arr.copy()
```

```python
pic_red[:, :, 1] = 0 #(163, 310) Zero out contribution from green
pic_red[:, :, 2] = 0 #(163, 310) Zero out contribution from blue
```

```python
plt.imshow(pic_red) # Isolated red images 
```

```python
pic_blue = pic_arr.copy()
```

```python
# isolated blue image 
pic_blue[:, :, 0] = 0 
pic_blue[:, :, 1] = 0 
```

```python
plt.imshow(pic_blue)
```
